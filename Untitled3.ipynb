```python
import os

# Nome do arquivo de repositório
ARQUIVO_REPOSITORIO = "produtos.txt"


def carregar_produtos():
    """Carrega os produtos do arquivo TXT."""
    if not os.path.exists(ARQUIVO_REPOSITORIO):
        return []
    with open(ARQUIVO_REPOSITORIO, "r") as arquivo:
        linhas = arquivo.readlines()
        return [eval(linha.strip()) for linha in linhas]


def salvar_produtos(produtos):
    """Salva os produtos no arquivo TXT."""
    with open(ARQUIVO_REPOSITORIO, "w") as arquivo:
        for produto in produtos:
            arquivo.write(f"{produto}\n")


def menu():
    """Exibe o menu do sistema."""
    print("\n=== Catálogo de Produtos ===")
    print("(1) Inclusão")
    print("(2) Alteração baseado em campo chave")
    print("(3) Exclusão em campo chave")
    print("(4) Relatório Geral")
    print("(5) Pesquisa Parcial por campo de relevância")
    print("(6) Saída")


def incluir_produto(produtos):
    """Inclui um novo produto."""
    codigo = input("Digite o código do produto (único): ").strip()
    if any(produto["codigo"] == codigo for produto in produtos):
        print("Erro: Código já cadastrado!")
        return produtos

    nome = input("Digite o nome do produto: ").strip()
    preco = float(input("Digite o preço do produto: "))
    categoria = input("Digite a categoria do produto: ").strip()

    novo_produto = {"codigo": codigo, "nome": nome, "preco": preco, "categoria": categoria}
    produtos.append(novo_produto)
    print("Produto adicionado com sucesso!")
    return produtos


def alterar_produto(produtos):
    """Altera um produto baseado no código."""
    codigo = input("Digite o código do produto a ser alterado: ").strip()
    for produto in produtos:
        if produto["codigo"] == codigo:
            print("Produto encontrado:")
            print(produto)
            produto["nome"] = input("Digite o novo nome: ").strip()
            produto["preco"] = float(input("Digite o novo preço: "))
            produto["categoria"] = input("Digite a nova categoria: ").strip()
            print("Produto alterado com sucesso!")
            return produtos
    print("Erro: Produto não encontrado!")
    return produtos


def excluir_produto(produtos):
    """Exclui um produto baseado no código."""
    codigo = input("Digite o código do produto a ser excluído: ").strip()
    produtos_filtrados = [produto for produto in produtos if produto["codigo"] != codigo]
    if len(produtos) == len(produtos_filtrados):
        print("Erro: Produto não encontrado!")
    else:
        print("Produto excluído com sucesso!")
    return produtos_filtrados


def relatorio_geral(produtos):
    """Exibe o relatório geral dos produtos."""
    print("\n=== Relatório Geral ===")
    if not produtos:
        print("Nenhum produto cadastrado.")
    else:
        for produto in produtos:
            print(f"Código: {produto['codigo']}, Nome: {produto['nome']}, "
                  f"Preço: R${produto['preco']:.2f}, Categoria: {produto['categoria']}")
    print("=" * 30)


def pesquisa_parcial(produtos):
    """Pesquisa parcial por nome."""
    termo = input("Digite o termo para pesquisa parcial: ").strip().lower()
    resultados = [produto for produto in produtos if termo in produto["nome"].lower()]
    if resultados:
        print("\n=== Resultados da Pesquisa ===")
        for produto in resultados:
            print(f"Código: {produto['codigo']}, Nome: {produto['nome']}, "
                  f"Preço: R${produto['preco']:.2f}, Categoria: {produto['categoria']}")
    else:
        print("Nenhum produto encontrado com o termo fornecido.")
    print("=" * 30)


def main():
    """Função principal."""
    produtos = carregar_produtos()

    while True:
        menu()
        opcao = input("Escolha uma opção: ").strip()

        if opcao == "1":
            produtos = incluir_produto(produtos)
        elif opcao == "2":
            produtos = alterar_produto(produtos)
        elif opcao == "3":
            produtos = excluir_produto(produtos)
        elif opcao == "4":
            relatorio_geral(produtos)
        elif opcao == "5":
            pesquisa_parcial(produtos)
        elif opcao == "6":
            salvar_produtos(produtos)
            print("Saindo... Dados salvos.")
            break
        else:
            print("Opção inválida. Tente novamente!")


if __name__ == "__main__":
    main()
```
